```r

library(ggplot2)
data <- data.frame(
  x = rnorm(100),
  y = rnorm(100)
)
ggplot(data, aes(x = x, y = y)) +
  geom_point(aes(size = abs(x - y), color = abs(x - y))) +
  scale_size_continuous(range = c(3, 10)) +
  scale_color_gradient(low = "blue", high = "red") +
  labs(title = "Exemple de visualisation non triviale", x = "X", y = "Y")
```
